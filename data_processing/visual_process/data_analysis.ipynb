```python
station_NYC = pd.read_csv('.\\m_station_NYC.csv', usecols=['station_name','station_lat','station_lng'])
weekday_name = ['Sun', 'Mon', 'Tues', 'Wed', 'Thur', 'Fri', 'Sat']

for i, w in enumerate(weekday_name):
    inflow_attribute = ('{}_inflow').format(weekday_name[(start_weekday+i)%7])
    outflow_attribute = ('{}_outflow').format(weekday_name[(start_weekday+i)%7])

    station_NYC[inflow_attribute] = np.zeros(station_NYC.shape[0])
    station_NYC[outflow_attribute] = np.zeros(station_NYC.shape[0])

for k in range(1,13):
    data = np.load('.\\processed_data\\2019{}_.npy'.format(str(k).zfill(2)))
    outflow = data[:,:,0].T
    inflow = data[:,:,1].T

    start_weekday = datetime.datetime(2019,k,1).weekday()
    weekday_list = []

    for j in range(7):
        weekday_list.append([i for i in range(outflow.shape[1]) if (i // 24) % 7 == j])

    for j in range(station_NYC.shape[0]):
        for i, w in enumerate(weekday_list):
            inflow_attribute = ('{}_inflow').format(weekday_name[(start_weekday+i)%7])
            outflow_attribute = ('{}_outflow').format(weekday_name[(start_weekday+i)%7])

            station_NYC.loc[j, inflow_attribute] += inflow[j, w].mean()
            station_NYC.loc[j, outflow_attribute] += outflow[j,w].mean()

station_NYC.head()
    
```
