```python
import pandas as pd
```

```python
import json
dataimport= json.load(open("data_bicycle3.json"))
dic={}
for line in dataimport.keys():
    dic[(float(line[line.index("(")+1:line.index(",")]),float(line[line.index(",")+1:line.index(")")]))]=tuple(dataimport[line])
print(len(dic))
```

```python
import numpy as np
def time_unit_transform(time_unit, raw_data, year,month, dic):
    day_month={1:31,2:28,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}
    if year==2020:
        day_month[2]=29

    room = np.full((day_month[month] * (1440 // time_unit), len(dic), 2), 0)
    for i in range(len(raw_data)):
        x = raw_data.loc[i, 'inflow'] // time_unit
        x2 = raw_data.loc[i, 'outflow'] // time_unit
        lat_end = raw_data.loc[i, 'end station latitude']
        lon_end = raw_data.loc[i, 'end station longitude']
        lat_start = raw_data.loc[i, 'start station latitude']
        lon_start = raw_data.loc[i, 'start station longitude']
        if (x>=(day_month[month]*(1440//time_unit))) or (x2>=(day_month[month]*(1440//time_unit))):
            continue
        if (lat_end, lon_end) in dic.keys():
            room[x][dic[(lat_end, lon_end)][1]][0] += 1
        if (lat_start, lon_start) in dic.keys():
            room[x2][dic[(lat_start, lon_start)][1]][1] += 1
    np.save(str(year)+str(month)+".npy", room)
```
